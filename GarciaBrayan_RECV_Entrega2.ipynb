```python
#Se cargan los documentos

import pandas as pd

data = pd.read_csv('Cars93.csv')
data.keys()
```

```python
#Se muestran los datos

data.head(10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

#--- Verificamos si existen celdas vacias
data.isnull().sum()
```

```python
# Colocamos en los nulos el medio
Rear=np.mean(data["Rear.seat.room"])
data["Rear.seat.room"][pd.isnull(data["Rear.seat.room"])]=Rear


Luggage=np.mean(data["Luggage.room"])
data["Luggage.room"][pd.isnull(data["Luggage.room"])]=Luggage
```

```python
#--- Verificamos si existen celdas vacias
data.isnull().sum()
```

```python
#Se toman las llaves ignorando las de precio
keys=[ 'Manufacturer','Type','MPG.city', 'MPG.highway', 'AirBags', 'DriveTrain',
       'Cylinders', 'EngineSize', 'Horsepower', 'RPM', 'Rev.per.mile',
       'Man.trans.avail', 'Fuel.tank.capacity', 'Passengers', 'Length',
       'Wheelbase', 'Width', 'Turn.circle', 'Rear.seat.room', 'Luggage.room',
       'Weight', 'Origin']
y=["Price"]
```

```python
#Se cambian los datos cualitativos a cuantitativos
kk,kl=np.unique(data['Manufacturer'], return_counts=True)
for i in range(len(data["Manufacturer"])):
  for j in range(len(kk)) :
    if(data["Manufacturer"][i]==kk[j]):
      data["Manufacturer"][i]=j
data["Manufacturer"]=data["Manufacturer"]      
```

```python
#Se cambian los datos cualitativos a cuantitativos
kk,kl=np.unique(data['DriveTrain'], return_counts=True)
for i in range(len(data["DriveTrain"])):
  for j in range(len(kk)) :
    if(data["DriveTrain"][i]==kk[j]):
      data["DriveTrain"][i]=j
data["DriveTrain"]=data["DriveTrain"]
```

```python
#Se cambian los datos cualitativos a cuantitativos
kk,kl=np.unique(data['Man.trans.avail'], return_counts=True)
for i in range(len(data["Man.trans.avail"])):
  for j in range(len(kk)) :
    if(data["Man.trans.avail"][i]==kk[j]):
      data["Man.trans.avail"][i]=j
data["Man.trans.avail"]=data["Man.trans.avail"]
```

```python
#Se cambian los datos cualitativos a cuantitativos
kk,kl=np.unique(data['Origin'], return_counts=True)
for i in range(len(data["Origin"])):
  for j in range(len(kk)) :
    if(data["Origin"][i]==kk[j]):
      data["Origin"][i]=j
data["Origin"]=data["Origin"]
```

```python
#Se visualiza los datos 
for k in keys:
  fig = plt.figure(figsize=(7,5))
  plt.plot(data[k],data[y],'*')
  plt.xlabel(k)
  plt.ylabel(y)
  plt.show()
```

```python
# Usamos corner y tranformamos los datos
import corner
from sklearn.preprocessing import StandardScaler

data_scaled = StandardScaler().fit_transform(data[keys])
figure = corner.corner(data[keys], labels=keys,
                       quantiles=[0.16, 0.5, 0.84],
                       show_titles=True, title_kwargs={"fontsize": 12})
```

```python
#Se convierte Y en binario
Y=data["Price"]

Y[Y<= 30]=0
Y[Y>30]=1
plt.plot(Y)
```

```python
#Se carga el modelo de arbol

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model
```

```python

# Entrenamos nuestro modelo

from sklearn.model_selection import train_test_split

model = RandomForestClassifier(n_estimators=100, max_depth=10)

X_train, X_test, Y_train, Y_test = train_test_split(np.array(data[keys]),Y)

model.fit(X_train,Y_train)

Y_pred = model.predict(X_test)
```

```python
#Se calcula la presicion
acc = sum(Y_pred == Y_test)/len(Y_test)
acc
```

```python
#--- Calculamos la importancia de los parametros
importances = model.feature_importances_
importances
```

```python
#--- Organizamos los parametros de mayor a menor
ii = np.argsort(importances)[::-1]
np.array(keys)[ii]
```
